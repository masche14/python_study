# 랜덤 조짜기 구현

```python
# 인원 번호 생성
org_list=[i for i in range(33)]

# 조 편성을 저장할 리스트
team_list=[]

# 조 편성을 위한 예비 리스트
for_sample_list=org_list[0:]

# 조당 인원수 입력
member_cnt=int(input('한 조당 인원수를 입력하세요 : '))

for i in range(int(len(for_sample_list)/member_cnt)):
    a=random.sample(for_sample_list, member_cnt)
    team_list.append(a)
    for i in a:
        for_sample_list.remove(i)

# 조 편성에서 제외된 인원 랜덤 팀에 한명씩 배정
if for_sample_list!=[]:
    
    index_list=[]

    for i in for_sample_list:
        except_one=i
        while True:
            plus_index=random.randint(0,len(team_list)-1)
            if plus_index not in index_list:
                index_list.append(plus_index)
                team_list[plus_index].append(except_one)
                break
    

print("각 조당 명단")
for i in range(len(team_list)):
    team=', '.join([str(i) for i in team_list[i]])
    print(f'{i+1}. {team}')
```
